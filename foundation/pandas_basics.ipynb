```python
# Preprocessing

df[pd.isnull(df).any(axis=1)].head()
```

```python
df = df.dropna()
df = df.reset_index()
df.head()
```

```python
df = df.drop(["name", "index", "cabin", "ticket"], axis=1)
df.head()
```

```python
df["sex"] = df["sex"].map({"female": 0, "male": 1}).astype("int")
df["embarked"] = df["embarked"].map({"S": 0, "C": 1, "Q": 2}).astype("int")
df.head()
```
